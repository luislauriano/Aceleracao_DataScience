##**Criando o modelo com Random Forest**

```
from sklearn.ensemble import RandomForestRegressor
```

```
modelo = RandomForestRegressor(n_estimators=100)
modelo.fit(x_treino, y_treino)
```

```

y_pred = modelo.predict(x_teste)
```

##**Dividir os dados em treino e teste e verificar a acuracia e precisão**

```
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```
_train, x_test, y_train, y_test = train_test_split(x_treino, y_treino, test_size=0.2, random_state=0)
```

```
y_pred_met = modelo.predict(x_test)
```

```
print('Explained Variance Score', metrics.explained_variance_score(y_test, y_pred_met)) 
#The best possible score is 1.0, lower values are worse.
print('Max Error:', (metrics.max_error(y_test, y_pred_met)))
# maior real - predito
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred_met))
# MAE = (real - predito) / n, erro médio
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred_met)))
# RMSE = raiz do quadrado do anterior
print('Median Squared Error:', metrics.median_absolute_error(y_test, y_pred_met))
# mediana dos (real - predito)
print('R² score:', metrics.r2_score(y_test, y_pred_met))
```
